为了分析主演的票房号召力，我们建立了一个数学模型。作为实际情况的简化，这里我们将票房的组成分为两部分，即主演的号召力和影片本身质量的吸引力，导演对影片的影响部分考虑进影片质量中去。  
主演的号召力我们这里用两个指标来衡量，即主演的**社交媒体影响力**和主演的**口碑**，体现在数据中即为相关的微博数量及其倾向。影片本身质量我们使用各个平台的评分加权得到（考虑到各个平台的偏向不同）。  
作为演示，我们这里将省去数据处理和特征提取的部分，专注于模型和算法。  
*细节见analy.py*

首先读取数据

```python
movie_comments = analy.get_movie_comments_handled()
box = analy.get_box_office()
```

这里我们使用了线性模型，考虑到实际情况我们对模型做了修改，也就是使用了NNLS模型（*详见nnls.py*），训练后得到了参数和偏置项。  
接下来我们将参数传入模型，得到修正后的票房影响数据。

```python
coef, intercept = analy.train()
box_influence = analy.estimate_actor_influence(movie_comments, coef[-1], intercept, box)
```

最后我们来观察一下结果。可以发现基本上还是符合直觉的——电影票房越高，主演粉丝群体越大，其票房号召力也越大。
